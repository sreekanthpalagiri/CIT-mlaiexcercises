```python
from collections import Counter
import os
import math
import re
import string
```

```python
print(os.getcwd())
os.chdir('C:\\Work\\CIT\\COMP9061-ML\\NaiveBayees\\')

pospath=os.getcwd()+'\\data\\train\\pos'
negpath=os.getcwd()+'\\data\\train\\neg'
print(pospath,negpath)

poslisting = os.listdir(pospath)
neglisting = os.listdir(negpath)
poswords=[]
negwords=[]

probpos = len(poslisting)/(len(poslisting)+len(neglisting))
probneg = len(neglisting)/(len(poslisting)+len(neglisting))
print('P(Pos):',probpos,'P(Neg):',probneg)
```

```python
for file in poslisting:
    f=open(pospath+'\\'+file,'r',encoding='utf8')
    data = f.read() 
    data = re.sub('<.*?>', ' ', data, flags=re.MULTILINE)
    data = re.sub('['+re.escape(string.punctuation)+']', '', data, flags=re.MULTILINE)
    allwords=data.split()
    f.close()
    allwords = [w.lower() for w in allwords]
    poswords = poswords + allwords
```

```python
print('Total words in pos reviews:',len(poswords))
#Total words in pos reviews: 2958832
#Total words in pos reviews: 2909538
```

```python
for file in neglisting:
    f=open(negpath+'\\'+file,'r',encoding='utf8')
    data = f.read() 
    data = re.sub('<.*?>', ' ', data, flags=re.MULTILINE)
    data = re.sub('['+re.escape(string.punctuation)+']', '', data, flags=re.MULTILINE)
    allwords=data.split()
    f.close()
    allwords = [w.lower() for w in allwords]
    negwords = negwords + allwords
   
print('Total words in neg reviews:',len(negwords))
#Total words in neg reviews: 2885848
#Total words in neg reviews: 2849994
```

```python
vocab=set(poswords+negwords)
print('Total unique words in vocabulary:',len(vocab))
#Total unique words in vocabulary: 251637
```

```python
posdict=dict.fromkeys(vocab,0)
negdict=dict.fromkeys(vocab,0)
for x in list(posdict)[0:5]:
    x=re.sub(re.escape('!"#$%&()*+,-./:;<=>?@[\]^_`{|}~'), '', x, flags=re.MULTILINE)
    x=re.sub(r'\S*@\S*\s?', '', x, flags=re.MULTILINE)
    print (x)
for x in list(negdict)[0:5]:
    print (x)
```

```python
poscounts = Counter(poswords)
negcounts = Counter(negwords)
print('Length of poscounts:',len(poscounts))
print('Length of negcounts:',len(negcounts))
```

```python
posdict.update(poscounts)
negdict.update(negcounts)
print('Length of posdict:',len(posdict))
for i in sorted(posdict)[:50]:
    print (i,'----', posdict[i])
print('Length of negdict:',len(posdict))
for i in sorted(negdict)[:50]:
    print (i, '----',negdict[i])
```

```python
probworkposdict={}
probwordnegdict={}
lenpos = len(poscounts)
lenneg = len(negcounts)
lenvocab= len(vocab)

for key in sorted(posdict):
    probworkposdict[key]= ((1 + posdict[key] )/(lenpos + lenvocab))
    
print('Length of probworkposdict:',len(probworkposdict))

for key in sorted(negdict):
    probwordnegdict[key]= ((1 + negdict[key] )/(lenneg + lenvocab))
    
print('Length of probwordnegdict:',len(probwordnegdict))
```

```python
print('probwordposdict')
for i in sorted(probworkposdict)[:10]:
    print (i,'----', probworkposdict[i])

print('probwordnegdict')
for i in sorted(probwordnegdict)[:10]:
    print (i, '----',probwordnegdict[i])
```

```python
postestpath=os.getcwd()+'\\data\\test\\pos'
negtestpath=os.getcwd()+'\\data\\test\\neg'
print(postestpath,negtestpath)

postestlisting = os.listdir(postestpath)
negtestlisting = os.listdir(negtestpath)
postestwords=[]
negtestwords=[]
```

```python
i=0
resultpos=[]
for file in postestlisting:
    f=open(postestpath+'\\'+file,'r',encoding='utf8')
    allwords=f.read().split()
    f.close()
    posprodtest= math.log(probpos)
    negprodtest= math.log(probneg)
    i+=1
    for word in allwords:
        if (word in probworkposdict.keys()):
            posprodtest= posprodtest + math.log(probworkposdict[word])
            negprodtest= negprodtest + math.log(probwordnegdict[word])
    resultpos.append(1) if posprodtest > negprodtest else resultpos.append(0)
    print(i,'Pos:',posprodtest,'Neg:',negprodtest)
```

```python
print('Accuracy Pos Test:',sum(resultpos)/len(resultpos))
```

```python
i=0
resultneg=[]
for file in negtestlisting:
    f=open(negtestpath+'\\'+file,'r',encoding='utf8')
    allwords=f.read().split()
    f.close()
    posprodtest= math.log(probpos)
    negprodtest= math.log(probneg)
    i+=1
    for word in allwords:
        if (word in probworkposdict.keys()):
            posprodtest += math.log(probworkposdict[word])
            negprodtest += math.log(probwordnegdict[word])
    resultneg.append(0) if posprodtest > negprodtest else resultneg.append(1)
    print(i,'Pos:',posprodtest,'Neg:',negprodtest)
```

```python
print('Accuracy Neg Test:',sum(resultneg)/len(resultneg))
```
